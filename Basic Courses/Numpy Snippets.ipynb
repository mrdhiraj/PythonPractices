```python
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt
aa = plt.subplot(1,2,1)
bb = plt.subplot(1,2,2)
#loc is mean , scale is standard deviation , size is sample size
x=np.random.normal(loc=0, scale=1, size=1500)#try changing size
aa.plot(x,'-')
bb.hist(x,bins=50,normed=True)#normed will make total area under histogram =1
print(x.mean(),x.std())
```

```python
import matplotlib.gridspec as gridspec
#gridspec.GridSpec?
plt.figure()
gspec = gridspec.GridSpec(3, 3)

top_histogram = plt.subplot(gspec[0, 1:])
side_histogram = plt.subplot(gspec[1:, 0])
lower_right = plt.subplot(gspec[1:, 1:])


Y = np.random.normal(loc=0.0, scale=1.0, size=10000)
X = np.random.random(size=10000)
lower_right.scatter(X, Y,label='Scatter')
top_histogram.hist(X, bins=100,label='Random Distribution',normed=True)
s = side_histogram.hist(Y, bins=100, orientation='horizontal' ,label ='Normal Distribution',normed=True)
side_histogram.invert_xaxis()
#plt.gcf?
for ax in plt.gcf().get_axes():
    ax.legend()
    for label in ax.get_xticklabels() + ax.get_yticklabels():
        label.set_visible(True)


# change axes limits make same
for ax in [top_histogram, lower_right]:
    ax.set_xlim(0, 1)
for ax in [side_histogram, lower_right]:
    ax.set_ylim(-5, 5)        
```
